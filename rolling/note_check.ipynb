```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp

import lsst.sims.maf.db as db
import lsst.sims.maf.utils as utils
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.stackers as stackers
import lsst.sims.maf.metricBundles as metricBundles
import lsst.sims.maf.plots as plots
```

```python
dbFile = 'rolling_mod6_sdf_0.10_v1.5_10yrs.db'
conn = db.OpsimDatabase(dbFile)
outDir='temp'
resultsDb = db.ResultsDb(outDir=outDir)
```

```python
bundleList = []
nights = 365./2. + np.arange(0,365*10, 365)
for night in nights:
    sql = 'note = "greedy" and filter="y" and night > %i and night < %i' % (night, night+365)

    plotDict = {'rot': (90, 90, 90), 'flip': 'geo', 'norm': 'log'}
    plotFunc = plots.HealpixSkyMap()

    metric = metrics.CountMetric(col='filter')
    slicer = slicers.HealpixSlicer(latCol='altitude', lonCol='azimuth', useCache=False)
    bundleList.append(metricBundles.MetricBundle(metric,slicer,sql, plotDict=plotDict, plotFuncs=[plotFunc]))
```

```python
bd = metricBundles.makeBundlesDictFromList(bundleList)
bg = metricBundles.MetricBundleGroup(bd, conn, outDir=outDir, resultsDb=resultsDb)
bg.runAll()
```

```python
bg.plotAll(closefigs=False)
```
